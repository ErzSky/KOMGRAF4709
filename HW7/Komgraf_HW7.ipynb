**Refleksi Terhadap Sumbu X**

```python
def refleksiSumbuX():
    matrix = np.array([[1, 0, 0],
                       [0, -1, 0],
                       [0, 0, 1]])
    return matrix

def apply_reflection(matrix, vector):
    resultVector = np.dot(matrix, vector)
    return resultVector

refleksiMatrix = refleksiSumbuX()

inputP1 = np.array([1, 3, 1])
inputP2 = np.array([6, 1, 1])

outputP1 = apply_reflection(refleksiMatrix, inputP1)
outputP2 = apply_reflection(refleksiMatrix, inputP2)

print("P1:", inputP1)
print("P2:", inputP2)
print("Refleksi P1:", outputP1)
print("Refleksi P2:", outputP2)
```

**Translasi**

```python
def translasi(matriks, tx, ty):
    translasiMatrix = np.array([[1, 0, tx],
                                [0, 1, ty],
                                [0, 0, 1]])
    hasil = np.dot(translasiMatrix, matriks)
    return hasil

def main():

    x = int(input("Masukkan nilai x : "))
    y = int(input("Masukkan nilai y : "))
    matriksAwal = np.array([[1, 0, x],
                            [0, 1, y],
                            [0, 0, 1]])

    print("Matriks awal:")
    print(matriksAwal)

    tx = int(input("Masukkan nilai tx untuk translasi: "))
    ty = int(input("Masukkan nilai ty untuk translasi: "))

    matriksHasil = translasi(matriksAwal, tx, ty)

    print("Matriks setelah translasi:")
    print(matriksHasil)

if __name__ == "__main__":
    main()
```

**Skalar**

```python
def skalar(matriks, sx, sy):
    skalarMatrix = np.array([[sx, 0, 0],
                             [0, sy, 0],
                             [0, 0, 1]])

    hasil = np.dot(skalarMatrix, matriks)

    return hasil

def main():
    matriksAwal = np.array([[1, 0, 1],
                            [0, 1, 3],
                            [0, 0, 1]])

    print("Matriks awal:")
    print(matriksAwal)

    skalarX = int(input("Masukkan nilai skalar sx: "))
    skalarY = int(input("Masukkan nilai skalar sy: "))

    hasilSkalar = skalar(matriksAwal, skalarX, skalarY)

    print("Matriks setelah operasi skalar:")
    for row in hasilSkalar:
        print([round(val, 2) for val in row])

if __name__ == "__main__":
    main()
```

**Rotasi**

```python
def rotasiTotal(matriks, sudut1, sudut2):
    totalRad = math.radians(sudut1 + sudut2)

    rotasiMatrix = np.array([[math.cos(totalRad), -math.sin(totalRad), 0],
                             [math.sin(totalRad), math.cos(totalRad), 0],
                             [0, 0, 1]])

    hasil = np.dot(rotasiMatrix, matriks)

    return hasil

def main():
    matriksAwal = np.array([[1, 0, 2],
                            [0, 1, 6],
                            [0, 0, 1]])

    print("Matriks awal:")
    print(matriksAwal)

    sudutTotal = float(input("Masukkan sudut rotasi total (dalam derajat): "))

    hasilRotasi = rotasiTotal(matriksAwal, sudutTotal, 0)

    print("Matriks setelah rotasi:")
    for row in hasilRotasi:
        print([round(val, 2) for val in row])

if __name__ == "__main__":
    main()
```

**Translasi + Rotasi**

```python
def translasi(matriks, tx, ty):
    translasiMatrix = np.array([[1, 0, tx],
                                [0, 1, ty],
                                [0, 0, 1]])
    hasil = np.dot(translasiMatrix, matriks)
    return hasil

def rotasiTotal(matriks, sudut1, sudut2):
    totalRad = math.radians(sudut1 + sudut2)

    rotasiMatrix = np.array([[math.cos(totalRad), -math.sin(totalRad), 0],
                             [math.sin(totalRad), math.cos(totalRad), 0],
                             [0, 0, 1]])

    hasil = np.dot(rotasiMatrix, matriks)

    return hasil

def main():
    x = int(input("Masukkan nilai x : "))
    y = int(input("Masukkan nilai y : "))
    matriksAwal = np.array([[1, 0, x],
                            [0, 1, y],
                            [0, 0, 1]])

    print("Matriks awal:")
    print(matriksAwal)

    tx = int(input("Masukkan nilai tx untuk translasi: "))
    ty = int(input("Masukkan nilai ty untuk translasi: "))

    sudutTotal = float(input("Masukkan sudut rotasi total (dalam derajat): "))

    matriksSetelahTranslasi = translasi(matriksAwal, tx, ty)

    hasilRotasi = rotasiTotal(matriksSetelahTranslasi, sudutTotal, 0)

    print("Matriks setelah translasi dan rotasi:")
    for row in hasilRotasi:
        print([round(val, 2) for val in row])

if __name__ == "__main__":
    main()
```
